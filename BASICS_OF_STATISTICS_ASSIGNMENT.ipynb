1.  Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss
nominal, ordinal, interval, and ratio scales.

ANS - Qualitative Data: This type of data describes qualities or characteristics. It is non-numerical and can be categorized based on attributes or properties. Examples include:

Eye color (e.g., blue, brown, green)

Types of cuisine (e.g., Italian, Chinese, Indian)

Customer feedback (e.g., satisfied, neutral, dissatisfied)

Quantitative Data: This type of data is numerical and can be measured or counted. It is further classified into discrete and continuous data. Examples include:

Number of students in a class (discrete)

Height of individuals (continuous)

Annual income (continuous)

Scales of Measurement:
Nominal Scale: This is the simplest form of measurement. It categorizes data without any order. Examples include:

Gender (e.g., male, female)

Marital status (e.g., single, married, divorced)

Blood type (e.g., A, B, AB, O)

Ordinal Scale: This scale categorizes data with a meaningful order but without a consistent difference between categories. Examples include:

Movie ratings (e.g., 1 star, 2 stars, 3 stars, etc.)

Education level (e.g., high school, bachelor's, master's, PhD)

Pain intensity (e.g., mild, moderate, severe)

Interval Scale: This scale categorizes data with meaningful order and consistent intervals between values, but it lacks a true zero point. Examples include:

Temperature in Celsius or Fahrenheit

IQ scores

Calendar years (e.g., 2000, 2001, 2002)

Ratio Scale: This scale has all the characteristics of the interval scale, plus a true zero point, which allows for meaningful comparisons using ratios. Examples include:

Weight (e.g., 0 kg, 50 kg, 100 kg)

Height (e.g., 0 cm, 150 cm, 180 cm)

Duration (e.g., 0 minutes, 30 minutes, 60 minutes)

4.  What is a box plot, and what can it tell you about the distribution of data?

ANS - A box plot, also known as a box-and-whisker plot, is a graphical representation used to visualize the distribution of a data set. It provides a summary of the data's central tendency, spread, and any potential outliers. Let's break down what a box plot can tell you:

Components of a Box Plot
Box:

Represents the interquartile range (IQR), which contains the middle 50% of the data.

The lower edge of the box is the first quartile (Q1), and the upper edge is the third quartile (Q3).

The line inside the box marks the median (Q2).

Whiskers:

Extend from the box to the smallest and largest values within 1.5 times the IQR from Q1 and Q3, respectively.

Represent the range of the data, excluding outliers.

Outliers:

Data points that fall outside the whiskers.

Typically indicated with dots or asterisks.

5. Discuss the role of random sampling in making inferences about populations.

ANS - Random sampling is a technique used to select a subset of individuals or observations from a larger population in such a way that each member of the population has an equal chance of being selected. This method ensures that the sample is representative of the population, allowing researchers to make valid inferences.

Importance in Making Inferences
Representation:

Random sampling helps ensure that the sample accurately reflects the characteristics of the population.

This representativeness minimizes bias and allows for generalization of the findings to the entire population.

Reduced Bias:

By giving each member an equal chance of being included, random sampling reduces selection bias.

This leads to more reliable and valid results.

Statistical Validity:

Random sampling forms the basis for many statistical analyses and tests.

It allows for the application of probability theory to make inferences about the population, such as estimating population parameters (e.g., mean, proportion) and testing hypotheses.

External Validity:

The findings from a random sample can be generalized to the larger population, enhancing the external validity of the research.

This is particularly important in fields like social sciences, medicine, and market research.

ANS- The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that is symmetric and bell-shaped. It is one of the most important and widely used distributions in statistics. Here are its key properties:

Symmetry:

The normal distribution is symmetric around its mean (Î¼). The left and right sides of the distribution are mirror images.

Mean, Median, and Mode:

In a normal distribution, the mean, median, and mode are all equal and located at the center of the distribution.

Bell-Shaped Curve:

The distribution forms a bell-shaped curve where most of the data points are clustered around the mean. The tails of the curve extend infinitely in both directions but never touch the horizontal axis.

Standard Deviation:

The spread of the distribution is determined by the standard deviation (Ïƒ). A smaller standard deviation indicates a narrower curve, while a larger standard deviation indicates a wider curve.

Asymptotic:

The tails of the normal distribution approach, but never touch, the horizontal axis. This means that extreme values are possible, but their probability decreases as they move further from the mean.

 11. Explain what a random variable is and differentiate between discrete and continuous random variables.


ANS - A random variable is a numerical value that represents the outcomes of a random phenomenon. It is a variable whose possible values are outcomes of a random process, and it can take on different values depending on the randomness inherent in the process.

Types of Random Variables
Random variables can be classified into two main types: discrete and continuous.

Discrete Random Variables
Definition: A discrete random variable can take on a countable number of distinct values. These values are often whole numbers and can be listed individually.

Examples:

Number of students in a classroom.

Number of heads in 10 coin tosses.

Number of defective items in a batch.

Characteristics:

The possible values can be counted.

The probability distribution of a discrete random variable is represented by a probability mass function (PMF).

Continuous Random Variables
Definition: A continuous random variable can take on any value within a given range. These values are not countable but can be measured with infinite precision.

Examples:

Height of individuals.

Time taken to run a marathon.

Temperature in a city.

Characteristics:

The possible values form a continuous interval.

The probability distribution of a continuous random variable is represented by a probability density function (PDF).
